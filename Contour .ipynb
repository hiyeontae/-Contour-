```python
import cv2
import matplotlib.pyplot as plt
from skimage import io

# Image Load
image = cv2.imread('C:/Users/Kimhyuntae/Desktop/grain_size/circle_DRX.JPG', cv2.IMREAD_COLOR)

# OpenCV
edges = cv2.Canny(image, 45, 100)

# view

image_with_edges = image.copy()
image_with_edges[edges != 0] = [0, 0, 0]  # 경계선 부분을 빨간색으로 표시


cv2.imwrite('C:/Users/Kimhyuntae/Desktop/grain_size/1.png', image_with_edges, [cv2.IMWRITE_PNG_COMPRESSION, 9])

# 결과 출력
plt.imshow(cv2.cvtColor(image_with_edges, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
